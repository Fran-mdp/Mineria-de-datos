Funcion para calcular la entropia

```python
def calcular_entropia(columna_objetivo):
    valores, cuentas = np.unique(columna_objetivo, return_counts=True)
    probabilidades = cuentas / cuentas.sum()
    entropia = -sum(probabilidades * np.log2(probabilidades))
    return entropia
```

Funcion para calcular la ganancia con respecto al target

```python
def calcular_ganancia_info(df, atributo, objetivo):
    entropia_total = calcular_entropia(df[objetivo])
    valores, cuentas = np.unique(df[atributo], return_counts=True)
    ponderado = sum(
        (cuentas[i] / np.sum(cuentas)) * calcular_entropia(df[df[atributo] == valores[i]][objetivo])
        for i in range(len(valores))
    )
    ganancia = entropia_total - ponderado
    return ganancia, entropia_total
```
